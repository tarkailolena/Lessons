```python
# №1 Из строки чисел с пробелами вернуть наибольшее и наименьшее число
def numb_string(numbers):
    nums = list(map(int, numbers.split()))
    max_num = nums[0]
    min_num = nums[0]
    for x in nums:
        if x > max_num:
            max_num = x
        if x< min_num:
            min_num = x
    print(f"{max_num},{min_num}")


numb_string("2 3 4 5 8 9")
```

```python
# №2 Допускается использование 4 или 6 цифр в пин-коде.

def pin(n):
    if not n.isdigit():
      return False
    if len(n) == 4 or len(n) == 6:
        return True
    else:
        return False
print(pin("2345"))
print(pin("1A34"))
print(pin("123456"))
```

```python
# №7 Вернуть из слова средний символ . Если длина слова нечетная = 1, если четная = 2
l = "Словcо"
def av_letter(l):
  if len(l) % 2 == 0:
    return l[(len(l)//2) - 1] + l[(len(l)//2)]
  else:
   if len(l) % 2 > 0:
       return l[len(l)//2]

l = "Лес"
l_1 = "Лесник"
l_2 = "Лесничество"
print(av_letter(l))
print(av_letter(l_1))
print(av_letter(l_2))
```

```python
# №8 Определить функцию, которая принимает целочисленный аргумент и возвращает логическое значение true, false в зависимости от того, евляется ли целое число простым
import math
def nmbr(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

print(nmbr(9))  
print(nmbr(4))   
print(nmbr(233))   
```

```python
#Напишите программу, которая с помощью встроенных функций map() и filter() удаляет из списка numbers все нечетные элементы, большие 47, а все четные элементы нацело делит на два (целочисленное деление – //).
#Полученные числа следует вывести на одной строке, разделив символом пробела и сохранив исходный порядок.

nums = [46, 61, 34, 17, 56, 26, 93, 1, 3, 82, 71, 0, 94, 89, 8, 36, 94, 46, 33]

res = list(map(lambda x: x // 2 if x % 2==0 else x, filter(lambda x: x % 2==0 or (x % 2 != 0 and x < 47), nums)))
print(res)
```
