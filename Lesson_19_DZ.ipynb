```python
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
X = df.drop('price_range', axis=1)
y = df['price_range']
print("Категории цен:", df['price_range'].unique())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = keras.Sequential([
    layers.Input(shape=(20,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(4, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Точность : {accuracy:.2%}")
print(f"Потери : {loss:.2%}")
```

```python
predict = model.predict(X_test[:1], verbose=0)
print(f"Предсказанная категория: {predict.argmax()}")
print(f"Реальная категория: {y_test.iloc[0]}")
```

```python
predict = model.predict(X_test[:10], verbose=0)
predict_classes = predict.argmax(axis=1)

print(f"Предсказанная для второго телефона: {predict_classes[1]}")
print(f"Реальная для второго телефона: {y_test.iloc[1]}")

print(f"Предсказаные для 10 телефонов: {predict_classes}")
print(f"Реальные для 10 телефонов: {y_test.values[:10]}")
```
