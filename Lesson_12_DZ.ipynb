```python
import sqlalchemy
```

```python
import sqlalchemy as db
```

```python
engine = db.create_engine('sqlite:///mydb8.db')
```

```python
conn = engine.connect()
```

```python
metadata = db.MetaData()
```

```python
dir(db)
```

```python
fruits = db.Table(
    'Fruits', metadata,
    db.Column('id', db.Integer),
    db.Column('name', db.Text),
    db.Column('cost', db.Integer),
    db.Column('color', db.Text)
)
fruits
```

```python
shop = db.Table(
    'Shop', metadata,
    db.Column('id', db.Integer),
    db.Column('id_fruits', db.Integer),
    db.Column('row', db.Integer),
    db.Column('promotion', db.Text)
)
shop
```

```python
metadata.create_all(engine)
```

```python
conn.execute(insertion) 
```

```python
insertion = shop.insert().values([
    {'id': '5', 'id_fruits': '2','row': 1 , 'promotion': 'yes'},
    {'id': '6', 'id_fruits': '3','row': 2 , 'promotion': 'no'},
    {'id': '7', 'id_fruits': '4','row': 3 , 'promotion': 'no'},
    {'id': '8', 'id_fruits': '4','row': 3 , 'promotion': 'no'}
])
```

```python
conn.execute(insertion) 
```

```python
select_all_query = db.select(shop)
select_result = conn.execute(select_all_query)
select_result.fetchall()
```

```python
join_query = db.select(
    fruits, shop.columns.promotion
).select_from(
    fruits.join(shop, shop.columns.id_fruits == fruits.columns.id)
)
```

```python
conn.execute(join_query) 
```

```python
select_result = conn.execute(join_query)
select_result.fetchall()
```

```python
join_query1 = db.select(
    fruits, shop
).join(shop, shop.columns.id_fruits == fruits.columns.id)
```

```python
conn.execute(join_query1) 
```

```python
select_result = conn.execute(join_query1)
select_result.fetchall()
```

```python
group = db.select(
    fruits.columns.name, 
    db.func.avg(fruits.columns.cost).label('cost_avg')
).group_by(
    fruits.columns.name 
)
```

```python
select_result = conn.execute(group)
select_result.fetchall()
```
