```python
class OneArray:
    def __init__(self, date):
        self.date = date
        self.size = (len(date),)
    def __str__(self):
        return f"{self.date}"
    def sum(self):
        return sum(self.date)
    def __add__(self, other):
        result = []
        for i in range(len(self.date)):
            sum_1 = self.date[i] + other.date[i]
            result.append(sum_1)
        return OneArray(result)
    def __mul__(self, other):
        result = [self.date[i] * other.date[i] for i in range(len(self.date))]
        return OneArray(result)

class TwoArray(OneArray):
    def __init__(self, date):
        super().__init__(date)
        self.row = len(date)
        self.col = len(date[0]) if self.row > 0 else 0
        self.date = date
    def __str__(self):
        return "\n".join([" ".join(map(str, row)) for row in self.date])
    def sum(self):
      flat_list = [num for row in self.date for num in row]
      return sum(flat_list)
    def __add__ (self, other):
        new_date = [[self.date[i][j] + other.date[i][j] for j in range(self.col)] for i in range(self.row)]
        return TwoArray(new_date)
    def __mul__(self, other):
       new_date = [[self.date[i][j] * other.date[i][j] for j in range(self.col)] for i in range(self.row)]
       return TwoArray(new_date)
    def find_saddle_points(self):
        saddle_points = []
        for i in range(self.row):
            for j in range(self.col):
                element = self.date[i][j]
                if element == min(self.date[i]):
                    column_values = [self.date[k][j] for k in range(self.row)]
                    if element == max(column_values):
                        saddle_points.append((i, j, element))  
        return saddle_points


class ThreeArray(TwoArray):
    def __init__(self, date):
        super().__init__(date)
        self.depth = len(date[0][0]) if len(date) > 0 and len(date[0]) > 0 else 0
        self.date = date
    def __str__(self):
     result = ""
     for matrix in self.date:
         result += "\n".join([" ".join(map(str, row)) for row in matrix])
         result += "\n\n"  # Разделитель между матрицами
     return result
    def __add__(self, other):
        new_date = [[[self.date[i][j][k] + other.date[i][j][k] for k in range(self.depth)] for j in range(self.col)] for i in range(self.row)]
        return ThreeArray(new_date)
    def sum(self):
     flat_list = [num for matrix in self.date for row in matrix for num in row]
     return sum(flat_list)
    def __mul__(self, other):
        new_date = [[[self.date[i][j][k] * other.date[i][j][k] for k in range(self.depth)] for j in range(self.col)] for i in range(self.row)]
        return ThreeArray(new_date)
```

```python
one_array_1 = OneArray([1, 2, 3])
one_array_2 = OneArray([4, 5, 6])
print(f" Первый массив:{one_array_1}")
print(f" Второй массив:{one_array_2}")
sum_elements2 = one_array_2.sum()
sum_elements1 = one_array_1.sum()
print(f" Сумма элементов второго массива: {sum_elements2}")
print(f" Сумма элементов первого массива: {sum_elements1}")
result_sum_array = one_array_1 + one_array_2
print(f"Сумма двух массивов: {result_sum_array}")
result_mul_array = one_array_1 + one_array_2
print(f"Умножение двух массивов: {result_sum_array}")
```

```python
two_array_1 = TwoArray([[1, 2] , [3, 4]])
two_array_2 = TwoArray([[5, 6], [7, 8]])
print(f"Первый двумерный массив: [{two_array_1}]")
print(f"Второй двумерный массив: [{two_array_2}]")
sum_two_array_1 = two_array_1.sum()
sum_two_array_2 = two_array_2.sum()
print(f"Сумма элементов второго массива: {sum_two_array_2}")
print(f"Сумма элементов первого массива: {sum_two_array_1}")
two_array_add = two_array_1 + two_array_2
print(f"Сумма двух двумерных массивов: [{two_array_add}")
result_mul_array_two = two_array_1 * two_array_2
print(f"Умножение двух массивов: {result_mul_array_two}")
saddle_points_1 = two_array_1.find_saddle_points()
print(f"Седловая точка в двумерном массиве_1: {saddle_points_1}")
saddle_points_2 = two_array_2.find_saddle_points()
print(f"Седловая точка в двумерном массиве_2: {saddle_points_2}")
```

```python
three_array_1 = ThreeArray([[[1, 2],[3, 4]],[[5, 6],[7, 8]]])
three_array_2 = ThreeArray([[[9, 10],[11, 12]],[[13, 14],[15, 16]]])
print(f"Первый трехмерный массив: {three_array_1}")
print(f"Второй трехмерный массив: {three_array_2}")
sum_three_array_1 = three_array_1.sum()
sum_three_array_2 = three_array_2.sum()
print(f"Сумма элементов второго массива: {sum_three_array_2}")
print(f"Сумма элементов первого массива: {sum_three_array_1}")
three_array_add = three_array_1 + three_array_2
print(f"Сумма двух трехмерных массивов: {three_array_add}")
result_mul_array_three = three_array_1 * three_array_2
print(f"Умножение двух массивов: {result_mul_array_three}")
saddle_points_3d = three_array_1.find_saddle_points()
print(f"Седловая точка в трехмерном массиве_1: {saddle_points_3d}")
saddle_points_3d = three_array_2.find_saddle_points()
print(f"Седловая точка в трехмерном массиве_2: {saddle_points_3d}")
```
