```python
# №1 Вычисления с матрицами
```

```python
import numpy as np
A = np.array([
    [1, 2, 4],
    [7, 5, -2],
    [-2, 1, 3]
])

B = np.array([
    [-1, 0, 3],
    [2, -6, 1],
    [-7, 1, 5]
])

C = np.array([
    [-2, 3],
    [4, 0],
    [7, -1]
])

result_sum = A + B

print("A + B:")
print(result_sum)

result_subtraction = B - A
print("B - A:")
print(result_subtraction)

result_mult_AC = A @ C 
print("A * C:")
print (result_mult_AC)

result_mult_ABC = A @ B @ C 
print("A * B * C:")
print(result_mult_ABC)
```
