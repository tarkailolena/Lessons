```python
plt.figure(figsize=(7, 5))
sns.histplot(df['medv']) 
plt.title('Распределение цен на жилье')
plt.xlabel('Цена в тыс. $')
plt.ylabel('Количество домов')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(y=df['age'])  
plt.title('Распределение возраста зданий ') 
plt.ylabel('Возраст здани ') 
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(y=df['crim'])  
plt.title('Бокс-плот уровня преступности')
plt.ylabel('Уровень преступности')
plt.show()
```

```python
df.plot.hexbin(x="dis", y="rm", gridsize=25);
plt.title('Диаграмма плотности')
plt.ylabel('Среднее количество комнат в жилище')
plt.xlabel('Взвешенные расстояния до центров занятости Бостона')
plt.show()
```

```python
ax = df.plot.scatter(x="dis", y="indus", color="DarkBlue", label="Group 1")

df.plot.scatter(x="rm", y="lstat", color="DarkGreen", label="Group 2", ax=ax);
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='rm', y='medv', data=df)
plt.title('Зависимость цены (MEDV) от среднего числа комнат (RM)')
plt.xlabel('Среднее число комнат (RM)')
plt.ylabel('Цена в тыс. $ (MEDV)')
plt.show()
```

```python
grouped_by_river = df.groupby('chas').mean()
print("Средние значения для домов: у реки (1) , не у реки (0):")
print(grouped_by_river[['medv', 'rm', 'lstat']])
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='chas', y='medv', data=df)
plt.title('Распределение цен в зависимости от близости к реке')
plt.xlabel('Рядом с рекой (0=Нет, 1=Да)')
plt.ylabel('Цена в тыс. $')
plt.show()
```

```python
df_min_crim = df[df['crim'] < 1]
df_max_crim = df[df['crim'] > 10]

plt.figure(figsize=(10, 6))
sns.kdeplot(df_min_crim['medv'], label='Низкий уровень преступности (<1)', fill=True, alpha=0.5)
sns.kdeplot(df_max_crim['medv'], label='Высокий уровень преступности (>10)', fill=True, alpha=0.5)

plt.title('Сравнение распределения цен жилья при разном уровне преступности')
plt.xlabel('Цена (тыс. $)')
plt.legend()
plt.show()
```
