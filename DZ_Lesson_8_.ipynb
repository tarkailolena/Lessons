```python
class Smartdevices:
    def __init__(self, colour, shape, brand):  
        self.__colour = colour
        self.__shape = shape
        self.__brand = brand
        
    def make_sound(self):  
        pass

    @property
    def __colour__(self):
        return self.__colour

    def display_colour(self):
        print(f"Colour: {self.__colour}")
        
    def __shape__(self):
        return self.__shape
        
    def display_shape(self):
        print(f"Shape: {self.__shape}")        
        
    def __brand__(self):
        return self.__brand
        
    def display_brand(self):
        print(f"Shape: {self.__brand}")  

    def display_info(self):
        print(f"colour:{self.__colour},shape: {self.__shape},brand: {self.__brand}")

class Cellphones(Smartdevices):
    def __init__(self, colour: str, shape: str, brand: str, diagonal: float):
        super().__init__(colour, shape, brand)  
        self.diagonal = diagonal
    def display_info(self):
        super().display_info()
        print(f"diagonal: {self.diagonal}")
    def make_sound(self):
        print("Dzzzz")

class Smartwatch(Smartdevices):
    def __init__(self, colour: str, shape: str, brand: str, steps: int):
        super().__init__(colour, shape, brand)
        self.steps = steps
    def display_info(self):
        super().display_info()
        print(f"steps: {self.steps}")
        

class Deviceshop:
    def __init__(self, name: str):
        self.list_devices = []
        self.name = name
    def add(self, device):
        self.list_devices.append(device)
    def date_info(self):
        for i in self.list_devices:
            i.display_info()

```

```python
cellphone = Cellphones(colour = "red", shape = "rectangle", brand = "Samsung", diagonal = 15)
cellphone.make_sound()    
cellphone.display_info()

smartwatch = Smartwatch(colour = "red", shape = "rectangle", brand = "Hoawei", steps = 2300)
smartwatch.display_info()

x = Deviceshop(name = "Q")
list_devices = [cellphone, smartwatch]
x.add(cellphone)
x.add(smartwatch)
x.date_info()
```

```python
import math

class Shape:
        def __init__(self, name = "Фигура"):  
            self._name = name
        def get_name(self):
            return self._name

        def area(self):
            return None      
        def perimeter(self):
            return None      
        def display_info(self):
            print(f"Name: {self.get_name()}")
            shape_area = self.area()
            if shape_area is not None:
                print(f"Area: {shape_area}")
            shape_perimeter = self.perimeter()
            if shape_perimeter is not None:
                print(f"Perimeter: {shape_perimeter}")    
                   
class Circle(Shape):
        def __init__(self, name:str, radius:int):
            super().__init__("Круг")
            if radius <= 0:
                return None
            self.radius = radius            
        def perimeter(self):
            return 2 * math.pi * self.radius       
        def area(self):
            return math.pi * (self.radius ** 2) 
            
            
class Rectangle(Shape):
        def __init__(self, name:str, widht:int , height:int):
            super().__init__("Прямоугольник")
            if widht <= 0 or height <= 0:
                return None
            self.widht = widht    
            self.height = height           
        def perimeter(self):
            return 2 * (self.height + self.widht)       
        def area(self):
            return self.height * self.widht   
            
class Triangle(Shape):
        def __init__(self, name:str, a:int, b:int, c:int):
            super().__init__("Треугольник")
            if a + b <= c or a + c <= b or b + c <= a:
                return None
            self.a = a   
            self.b = b
            self.c = c
        def perimeter(self):
            return self.a + self.b + self.c      
        def area(self):
            s = (self.a + self.b + self.c) / 2  
            return math.sqrt(s*(s-a)*(s-b)*(s-c))    
class Square(Rectangle):
        def __init__(self, lenght:int ):
            super().__init__("Квадрат", widht = lenght, height = lenght)
            self.lenght = lenght
            self.name = "Квадрат"

            
```

```python
radius = 3
circle = Circle(name = "Круг", radius = 3)
print(f"Radius:{radius}")
circle.display_info()
```

```python
class Employee:
    def __init__(self, id_e:int, name:str, position:str, hiring_date):
        self.id_e = id_e
        self.name = name
        self.position = position
        self.hiring_date = hiring_date
    def employee_info(self) -> float:
        return(f"ID: {self.id_e}, Имя: {self.name}, Должность: {self.position}, Дата приема: {self.hiring_date}") 
    def calculate_salary(self)->float:
        pass
    def calculate_vacation_pay(self) -> float:
        pass    

class FullTimeEmployee(Employee):
    def __init__(self, id_e:int, name:str, position:str, hiring_date, month_salary:float, med_insurance, bonus_rate):
        super().__init__(id_e, name, position, hiring_date)
        self.month_salary = month_salary
        self.med_insurance = 0.2
        self.bonus_rate = 0.01
    def calculate_salary(self) -> float:
        print(f"Месячная зп: {self.month_salary}")
    def calculat_vacation_pay(self) -> float:
         print(f"Сумма отпускных: {self.month_salary / 29 * 22}")
    def calculate_bonus_rate(self) -> float:
        print(f"Сумма бонусов: {self.month_salary  * self.bonus_rate}")
    def calculate_med_insurance(self) -> float:
        print(f"Сумма мед.страховки: {self.month_salary * 0.2}")
    def employee_info(self):
        base_info = super().employee_info()
        print(f" {base_info} , ЗП: {self.month_salary:}, Мед.страховка : {self.med_insurance:}, Бонусы:{self.bonus_rate:}")   

class PartTimeEmployee(Employee):
    def __init__(self, id_e:int, name:str, position:str, hiring_date, hourly_pay:float, hours_worked, limited_benefits):
        super().__init__(id_e, name, position, hiring_date)
        self.hourly_pay = 110
        self.hours_worked = 20
        self.limited_benefits = True
    def calculate_salary(self) -> float:
        print(f"Месячная зп: {self.hourly_pay * self.hours_worked}")
    def calculate_vacation_rate(self) -> float:
        print(f"Сумма отпускных: {(self.hourly_pay * self.hours_worked)/ 29 * 22 }")   
    def employee_info(self):
        base_info = super().employee_info()
        print(f" {base_info} , Отработанные часы: {self.hourly_pay}, Ставка за 1ч:{self.hours_worked}, Ограниченный соцпакет:{self.limited_benefits}")   

class CommissionEmployee(Employee):
    def __init__(self, id_e:int, name:str, position:str, hiring_date,  base_salary, protsent_of_prodazh, sales_value, bonus):
        super().__init__(id_e, name, position, hiring_date)
        self.base_salary = base_salary
        self.protsent_of_prodazh = protsent_of_prodazh
        self.sales_value = 10
        self.bonus = bonus
    def calculate_salary(self) -> float:
        print(f"Месячная зп: {self.base_salary + (self.sales_value * self.protsent_of_prodazh) + (self.base_salary * self.bonus)}")
    def calculate_vacation_rate(self) -> float:
        print(f"Сумма отпускных: {(self.base_salary + (self.sales_value * self.protsent_of_prodazh) / 29 * 22) }")   
    def employee_info(self):
        base_info = super().employee_info()
        print(f" {base_info} , Базовая ставка:{self.base_salary}, Процент от продаж:{self.protsent_of_prodazh}, Объем продаж: {self.sales_value}, Бонусы:{self.bonus}")    
class Manager(FullTimeEmployee): 
    def __init__(self, id_e:int, name:str, position:str, hiring_date, month_salary:float, med_insurance, bonus_rate, team_budget):
        super().__init__(id_e, name, position, hiring_date, month_salary, med_insurance, bonus_rate) 
        self._bonus_rate = 1000
        self.team_budget = team_budget
    def calculate_salary(self) -> float:
        base_salary = super().calculate_salary()
        print(f"Месячная зп + бонусы : {self.month_salary + self._bonus_rate}")
    def employee_info(self):
        base_info = super().employee_info()
        print(f" {base_info} , Бюджет на обучение: {self.team_budget:}")   
    
```
