```python
class Money:
    def __init__(self, rubles:int, kop:int):
        self.__rubles = rubles
        self.__kop = kop

    def sum(self, rubles, kop):
        self.__rubles += rubles
        self.__kop += kop
        if self.__kop > 99:
            self.__rubles += self.__kop // 100
            self.__kop = self.__kop % 100
        print(f"Money:{self.__rubles},{self.__kop}")

    def minus(self, rubles, kop):
        self.__rubles -= rubles
        self.__kop -= kop
        if self.__rubles + self.__kop //100  < 0:
            print("Нет средств")
        else:
            return(self.__rubles + self.__kop / 100)

    def umnog(self, rubles, kop):
        self.__rubles *= rubles
        self.__kop *= kop
        return(self.__rubles + self.__kop / 100)

    def delen(self, rubles, kop):
        self.__rubles /= rubles
        self.__kop /= kop
        return(self.__rubles + self.__kop / 100)

    def contrast(self, rubles, kop):
        if self.__rubles + self.__kop/100 >  rubles + kop/100:
          print("Денег на карте было больше,чем поступило")
        if self.__rubles + self.__kop/100 <  rubles + kop/100:
          print("Денег на карте было меньше,чем поступило")
        else:
          print("Поступившая сумма = сумме на карте")
```

```python
result = Money(25,10)
result.sum(10, 5)
```

```python
result_2 = Money(25, 10)
result_2.minus(20, 28)
```

```python
result_3 = Money(25,10)
result_3.umnog(3, 3)
```

```python
result_4 = Money(25, 10)
result_4.delen(5, 2)
```

```python
result_6 = Money(25,10)
result_6.contrast(25,40)
```

```python
class Matrix:
    def __init__(self, matrix):
        self.matrix = matrix
        self.rows = len(matrix)
        self.cols = len(matrix[0]) if self.rows > 0 else 0 

    def __str__(self):
        return "\n ".join([" ".join(map(str, row)) for row in self.matrix]) 

    def umnoj_skalar(self, scalar):
        result_matrix = []
        for i in range(self.rows):
            new_row = []
            for j in range(self.cols):
                new_row.append(self.matrix[i][j] * scalar)
            result_matrix.append(new_row)
        return Matrix(result_matrix)   

    def __add__ (self, other):
        new_matrix = [[self.matrix[i][j] + other.matrix[i][j] for j in range(self.cols)] for i in range(self.rows)]
        return Matrix(new_matrix)              
```

```python
matrix_1 = Matrix([[1, 2], [3, 4]])
matrix_2 = Matrix([[5, 6], [7, 8]])
```

```python
matrix_scalar_1 = matrix_1.umnoj_skalar(2) 
print(matrix_scalar_1)
```

```python
matrix_scalar_2 = matrix_2.umnoj_skalar(2) 
print(matrix_scalar_2)
```

```python
matrix_add = matrix_1 + matrix_2
print(matrix_add)
```

```python
class Fruit:
  def __init__(self, name, color, weight, price):
    self.name = name  #Инкапсуляция
    self.color = color
    self.weight = weight
    self.price = price

  def cost(self, weight, price):
    self.weight = weight
    self.price = price
    if self.weight > 0:
      return (f"К оплате: {self.weight * self.price}")

  def display_info(self):
     print(f"Фрукт: {self.name}")
     print(f"Цвет: {self.color} ")
     print(f"Вес: {self.weight} г")
     print(f"Цена: {self.price}")

class Citrus(Fruit): #Наследование
  def __init__(self, name, color, weight, price, level_vit_c):
    super().__init__(name, color, weight, price)
    self.level_vit_c = level_vit_c
  def display_info(self):
    super().display_info() #Полиморфизм
    print(f"Витамин С из {self.name}  равен {self.level_vit_c}")

class Apple(Citrus):
  def __init__(self, name, color, weight , price,  level_vit_c, sort):
    super().__init__(name, color, price,  weight,  level_vit_c,)
    self.sort = sort
  def display_info(self):
    super().display_info() #Полиморфизм
    print(f"Сорт {self.name} {self.sort}")

class Berries(Apple):
  def __init__(self, name, color, weight , price, level_vit_c, sort, ripe ):
    super().__init__(name, color, price, weight , level_vit_c, sort)
    self.ripe = ripe
  def display_info(self):
    super().display_info() #Полиморфизм
    print(f"Витамин С из {self.name}  равен {self.level_vit_c}")
```

```python
orange = Citrus("Апельсин", "Оранжевый", 450 , 20,  70 )
orange.display_info()
orange.cost(450, 20)
```

```python
apple = Apple("Яблоко", "Зеленое", 180 , 35, 40,  "Гренни" )
apple.display_info()
apple.cost(35,40)
```

```python
cherry = Berries("Вишня", "Красная", 1200 , 14, 5,  "Медовая", "Свежие" )
cherry.display_info()
cherry.cost(35,40)
```
